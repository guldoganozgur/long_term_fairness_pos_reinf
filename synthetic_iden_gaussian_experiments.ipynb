```python
import pickle
from utils import *

N = 400
c = np.array([0.1, 0.05, 0.2])
alpha = 0.4
theta_init = 0.25
lambda_ = 0.75
eta = 0.5
role_model = 0.5

mean = np.array([5,5]) # Groups U, V
std = np.array([1,1]) # Groups U, V
method = 'approx'
num_instances = 200
num_rounds = 40

decay = 0
seed_val = 23
```

```python
# Choice of expts to run:
pure_pos_reinf = True
role_model_reinf = True
weight_reinf = True
order_reinf = True

results = {}
```

```python
if pure_pos_reinf:
    np.random.seed(seed_val)
    applicants_u_MFG, admitted_u_MFG, theta_vec_MFG, role_model_MFG = process_MFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='positive')
    np.random.seed(seed_val)
    applicants_u_CMFG, admitted_u_CMFG, theta_vec_CMFG, role_model_CMFG = process_CMFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='positive')
    results['pure_pos_reinf'] = {'applicants_u_MFG': applicants_u_MFG, 'admitted_u_MFG': admitted_u_MFG, 'theta_vec_MFG': theta_vec_MFG, 'role_model_MFG': role_model_MFG, 'applicants_u_CMFG': applicants_u_CMFG, 'admitted_u_CMFG': admitted_u_CMFG, 'theta_vec_CMFG': theta_vec_CMFG, 'role_model_CMFG': role_model_CMFG}

if role_model_reinf:
    np.random.seed(seed_val)
    applicants_u_MFG, admitted_u_MFG, theta_vec_MFG, role_model_MFG = process_MFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, role_model_ratio=role_model, reinf_method='role_model')
    np.random.seed(seed_val)
    applicants_u_CMFG, admitted_u_CMFG, theta_vec_CMFG, role_model_CMFG = process_CMFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds+40, decay=decay, epsilon=0.01, role_model_ratio=role_model, reinf_method='role_model')
    results['role_model_reinf'] = {'applicants_u_MFG': applicants_u_MFG, 'admitted_u_MFG': admitted_u_MFG, 'theta_vec_MFG': theta_vec_MFG, 'role_model_MFG': role_model_MFG, 'applicants_u_CMFG': applicants_u_CMFG, 'admitted_u_CMFG': admitted_u_CMFG, 'theta_vec_CMFG': theta_vec_CMFG, 'role_model_CMFG': role_model_CMFG}

if weight_reinf:
    np.random.seed(seed_val)
    applicants_u_MFG, admitted_u_MFG, theta_vec_MFG, role_model_MFG = process_MFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='weight', weights=np.ones_like(c))
    np.random.seed(seed_val)
    applicants_u_CMFG, admitted_u_CMFG, theta_vec_CMFG, role_model_CMFG = process_CMFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='weight', weights=np.ones_like(c))
    results['weight_reinf'] = {'applicants_u_MFG': applicants_u_MFG, 'admitted_u_MFG': admitted_u_MFG, 'theta_vec_MFG': theta_vec_MFG, 'role_model_MFG': role_model_MFG, 'applicants_u_CMFG': applicants_u_CMFG, 'admitted_u_CMFG': admitted_u_CMFG, 'theta_vec_CMFG': theta_vec_CMFG, 'role_model_CMFG': role_model_CMFG}

if order_reinf:
    np.random.seed(seed_val)
    applicants_u_MFG, admitted_u_MFG, theta_vec_MFG, role_model_MFG = process_MFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='order', order=0.8)
    np.random.seed(seed_val)
    applicants_u_CMFG, admitted_u_CMFG, theta_vec_CMFG, role_model_CMFG = process_CMFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='order', order=0.8)
    results['order_reinf_fast'] = {'applicants_u_MFG': applicants_u_MFG, 'admitted_u_MFG': admitted_u_MFG, 'theta_vec_MFG': theta_vec_MFG, 'role_model_MFG': role_model_MFG, 'applicants_u_CMFG': applicants_u_CMFG, 'admitted_u_CMFG': admitted_u_CMFG, 'theta_vec_CMFG': theta_vec_CMFG, 'role_model_CMFG': role_model_CMFG}
    np.random.seed(seed_val)
    applicants_u_MFG, admitted_u_MFG, theta_vec_MFG, role_model_MFG = process_MFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='order', order=1.2)
    np.random.seed(seed_val)
    applicants_u_CMFG, admitted_u_CMFG, theta_vec_CMFG, role_model_CMFG = process_CMFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='order', order=1.2)
    results['order_reinf_slow'] = {'applicants_u_MFG': applicants_u_MFG, 'admitted_u_MFG': admitted_u_MFG, 'theta_vec_MFG': theta_vec_MFG, 'role_model_MFG': role_model_MFG, 'applicants_u_CMFG': applicants_u_CMFG, 'admitted_u_CMFG': admitted_u_CMFG, 'theta_vec_CMFG': theta_vec_CMFG, 'role_model_CMFG': role_model_MFG}
    
```

```python
lambdas = [0, 0.75, 3, 15, 100]

theta_vec_MFG_lambdas = []

for lambda_ in lambdas:
    np.random.seed(seed_val)
    applicants_u_MFG, admitted_u_MFG, theta_vec_MFG, role_model_MFG = process_MFG(N, c, alpha, mean, std, method, theta_init, lambda_, eta, num_instances, num_rounds, decay=decay, epsilon=0.01, reinf_method='positive')
    theta_vec_MFG_lambdas.append(theta_vec_MFG)

results['lambdas'] = {'lambdas': lambdas, 'theta_vec_MFG_lambdas': theta_vec_MFG_lambdas}
```

```python
percentiles_u_pure_pos = np.zeros([3,num_rounds])
percentiles_u_decresing_lambda = np.zeros([3,num_rounds+60])

percentiles_v_pure_pos = np.zeros([3,num_rounds])
percentiles_v_decresing_lambda = np.zeros([3,num_rounds+60])

for k in range(3):
    if k==0:
        percentiles_u_pure_pos[k,:] = 1-c[k]*results['pure_pos_reinf']['admitted_u_MFG'][k,:]/results['pure_pos_reinf']['applicants_u_MFG'][k,:]
        percentiles_v_pure_pos[k,:] = 1-c[k]*(1-results['pure_pos_reinf']['admitted_u_MFG'][k,:])/(1-results['pure_pos_reinf']['applicants_u_MFG'][k,:])

        percentiles_u_decresing_lambda[k,:] = 1-c[k]*results['pure_pos_reinf_decreasing_lambda']['admitted_u_MFG'][k,:]/results['pure_pos_reinf_decreasing_lambda']['applicants_u_MFG'][k,:]
        percentiles_v_decresing_lambda[k,:] = 1-c[k]*(1-results['pure_pos_reinf_decreasing_lambda']['admitted_u_MFG'][k,:])/(1-results['pure_pos_reinf_decreasing_lambda']['applicants_u_MFG'][k,:])

    else:
        percentiles_u_pure_pos[k,:] = percentiles_u_pure_pos[k-1,:] - c[k]*results['pure_pos_reinf']['admitted_u_MFG'][k,:]/results['pure_pos_reinf']['applicants_u_MFG'][k,:]
        percentiles_v_pure_pos[k,:] = percentiles_v_pure_pos[k-1,:] - c[k]*(1-results['pure_pos_reinf']['admitted_u_MFG'][k,:])/(1-results['pure_pos_reinf']['applicants_u_MFG'][k,:])

        percentiles_u_decresing_lambda[k,:] = percentiles_u_decresing_lambda[k-1,:] - c[k]*results['pure_pos_reinf_decreasing_lambda']['admitted_u_MFG'][k,:]/results['pure_pos_reinf_decreasing_lambda']['applicants_u_MFG'][k,:]
        percentiles_v_decresing_lambda[k,:] = percentiles_v_decresing_lambda[k-1,:] - c[k]*(1-results['pure_pos_reinf_decreasing_lambda']['admitted_u_MFG'][k,:])/(1-results['pure_pos_reinf_decreasing_lambda']['applicants_u_MFG'][k,:])

    
```

```python
fig_percentiles = plot_results_percentiles(percentiles_u_pure_pos, percentiles_v_pure_pos,'Percentiles for Pure Positive Reinforcement')
ax = fig_percentiles.gca()
ax.set_ylim([55,95])
fig_percentiles.savefig('figures/percentiles_pure_pos_reinf_MFG.pdf',bbox_inches='tight')
```
